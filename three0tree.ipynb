```python
# prompt: uma espiral de crescimento 3 no meu display

import numpy as np
import matplotlib.pyplot as plt

# Define the growth factor
growth_factor = 3

# Define the range for the angle (in radians)
theta = np.linspace(0, 8 * np.pi, 1000) # Go around 4 times

# Calculate the radius for each angle using the formula for a logarithmic spiral
# r = a * b^theta, here we can use r = a * growth_factor^(theta / (2 * pi))
# where 'a' is a scaling factor, let's use 1 for simplicity
a = 3
r = a * growth_factor**(theta / (2 * np.pi))

# Convert polar coordinates (r, theta) to Cartesian coordinates (x, y)
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create the plot
plt.figure(figsize=(8, 8))
plt.plot(x, y)

# Set plot title and labels
plt.title(f'A ESPIRAL {growth_factor}', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

# Set aspect ratio to be equal to avoid distortion
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.grid(True)
plt.show()
```
